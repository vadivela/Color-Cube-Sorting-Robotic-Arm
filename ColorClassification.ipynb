```python
import pandas as pd
import numpy as np
DATA = pd.read_csv(r"Path_to_Dataset")
y_raw = list(DATA['Y'])
size = 4056
Y=[[0]*4 for _ in range(size)]
X=[[0]*3 for _ in range(size)]
R = list(DATA['R'])
G = list(DATA['G'])
B = list(DATA['B'])
for i in range(size):
    Y[i][y_raw[i]] = 1
    X[i][0] = R[i]
    X[i][1] = G[i]
    X[i][2] = B[i]
X=np.array(X)
Y=np.array(Y)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
model.add(Dense(4,activation='softmax'))
model.compile(optimizer = tf.optimizers.Adam(),loss=tf.keras.losses.SparseCategoricalCrossentropy())
```

```python
model.fit(X,np.array(y_raw),epochs=1000)
```
